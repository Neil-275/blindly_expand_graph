```python
import pickle as pkl
import re
import os
from utils import extract_answer
```

```python
with open("knowledge_graph/KG_data/FB15k-237-betae/id2ent.pkl", "rb") as f:
        id2ent = pkl.load(f)
with open("knowledge_graph/KG_data/FB15k-237-betae/FB15k_mid2name.txt", "r", encoding='utf-8') as f:
        ent2name = {}
        for line in f:
            if line.strip():  # Skip empty lines
                mid, name = line.strip().split("\t", 1)
                ent2name[mid] = name
```

```python
preds = []
ground_truths = []
named_cands = []

for f in os.listdir('results/full'):
    with open(os.path.join('results/full', f), 'rb') as file:
        data = pkl.load(file)
        # print(data.keys())
        predi= data['predctions']
        gt= data['ground_truths']
        if len(gt)>len(predi):
            gt = gt[:-1]
        preds.extend(predi)
        ground_truths.extend(gt)
        for cand in data['cands']:
            named_cands.append([ent2name[id2ent[a]] for a in cand])
```
